```python
import time
```

```python
def unixtime(a):
    list = []
    for element in a:
        b = time.mktime(element.timetuple())
        list.append(b)
    return list
```

```python
L1_time_tick_FB = pd.to_datetime(Duxton_rent_final_L1_FB['Lease Start'])
L1_time_FB = unixtime(list(L1_time_tick_FB))
L1_time_tick_office = pd.to_datetime(Duxton_rent_final_L1_office['Lease Start'])
L1_time_office = unixtime(list(L1_time_tick_office))
L2_time_tick = pd.to_datetime(Duxton_rent_final_L2_office['Lease Start'])
L2_time = unixtime(list(L2_time_tick))
L3_time_tick = pd.to_datetime(Duxton_rent_final_L3['Lease Start'])
L3_time = unixtime(list(L3_time_tick))
L4_time_tick = pd.to_datetime(Duxton_rent_final_L4['Lease Start'])
L4_time = unixtime(list(L4_time_tick))
```

```python
L1_time_office
```

```python
import statsmodels.formula.api as smf
```

```python
data_1 = {'x': L1_time_FB,
         'y': list(Duxton_rent_final_L1_FB['Rent per sqft'])}
data_2 = {'x': L2_time,
         'y': list(Duxton_rent_final_L2_office['Rent per sqft'])}
data_3 = {'x': L3_time,
         'y': list(Duxton_rent_final_L3['Rent per sqft'])}
data_4 = {'x': L4_time,
         'y': list(Duxton_rent_final_L4['Rent per sqft'])}
data_5 = {'x': L1_time_office,
         'y': list(Duxton_rent_final_L1_office['Rent per sqft'])}
```

```python
model_1 = smf.ols('y ~ x', data = data_1).fit()
model_2 = smf.ols('y ~ x', data = data_2).fit()
model_3 = smf.ols('y ~ x', data = data_3).fit()
model_4 = smf.ols('y ~ x', data = data_4).fit()
model_5 = smf.ols('y ~ x', data = data_5).fit()
```

```python
L2_time.sort()
```

```python
L2_time_tick.sort_values()
```

```python
pd.qcut(L2_time_tick, 5).unique()
```

```python
tick = (L2_time[len(L2_time)-1] - L2_time[0]) / 5
```

```python
tick_list = np.arange(L2_time[0], L2_time[len(L2_time)-1], tick)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
plot_1 = [model_1.params[1] * x + model_1.params[0] for x in L1_time_FB]
plot_2 = [model_2.params[1] * x + model_2.params[0] for x in L2_time]
plot_3 = [model_3.params[1] * x + model_3.params[0] for x in L3_time]
plot_4 = [model_4.params[1] * x + model_4.params[0] for x in L4_time]
plot_5 = [model_5.params[1] * x + model_5.params[0] for x in L1_time_office]
plt.plot(L1_time_FB,plot_1, label = "Level 1 F&B rent", color = 'blue')
plt.plot(L2_time,plot_2, label = 'Level 2 rent', color = 'orange')
plt.plot(L3_time,plot_3, label = 'Level 3 rent', color = 'green')
plt.plot(L4_time,plot_4, label = 'Level 4 rent', color = 'red')
plt.plot(L1_time_office,plot_5, label = 'Level 1 office rent', color = 'yellow')
plt.scatter(L1_time_FB, list(Duxton_rent_final_L1_FB['Rent per sqft']), color = 'blue')
plt.scatter(L2_time, list(Duxton_rent_final_L2_office['Rent per sqft']), color = 'orange')
plt.scatter(L3_time, list(Duxton_rent_final_L3['Rent per sqft']), color = 'green')
plt.scatter(L4_time, list(Duxton_rent_final_L4['Rent per sqft']), color = 'red')
plt.scatter(L1_time_office, list(Duxton_rent_final_L1_office['Rent per sqft']), color = 'yellow')
ax.set_xticks(tick_list)
ax.set_xticklabels(['2020/2', '2021/7', '2022/10', '2023/2', '2023/5' ])
plt.title('Duxton Road Rental Trend')
plt.legend()
plt.savefig('Duxton Road Rental Trend')
```
